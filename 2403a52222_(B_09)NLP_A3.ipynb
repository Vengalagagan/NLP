Import NLTK

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
sentence = "Students are learning Natural Language Processing"
tokens = nltk.word_tokenize(sentence)
pos_tags = nltk.pos_tag(tokens)
```

```python
print(pos_tags)
```

Import Spacy

```python
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp("Students are learning Natural Language Processing")
for token in doc:
    print(token.text, token.pos_)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp("Apple is looking at buying a startup in India.")
for token in doc:
  print(token.text, token.pos_, token.tag_)
```

```python
import spacy
from collections import Counter
nlp = spacy.load("en_core_web_sm")
text = "Loving the new AI features ğŸ˜ #AI #MachineLearning"
doc = nlp(text)
nouns = []
verbs = []
for token in doc:
    if token.pos_ in ["NOUN", "PROPN"]:
        nouns.append(token.text)
    elif token.pos_ == "VERB":
        verbs.append(token.text)
noun_freq = Counter(nouns)
verb_freq = Counter(verbs)
print("Noun Frequency:", noun_freq)
print("Verb Frequency:", verb_freq)
```

Aim: Understand POS tagging challenges in informal, noisy text.

```python
import json

with open('enriched_posts.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

texts = [item['text'].encode('utf-8', errors='ignore').decode('utf-8') for item in data]

last_txt = " ".join(texts)
print(last_txt)
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
sentence = last_txt
tokens = nltk.word_tokenize(sentence)
pos_tags = nltk.pos_tag(tokens)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp(last_txt)
for token in doc:
  print(token.text, token.pos_, token.tag_)
```

```python
import spacy
from collections import Counter
nlp = spacy.load("en_core_web_sm")
text = last_txt
doc = nlp(text)
nouns = []
verbs = []
for token in doc:
    if token.pos_ in ["NOUN", "PROPN"]:
        nouns.append(token.text)
    elif token.pos_ == "VERB":
        verbs.append(token.text)
noun_freq = Counter(nouns)
verb_freq = Counter(verbs)
print("Noun Frequency:", noun_freq)
print("Verb Frequency:", verb_freq)
```
