Cosine Similarity

```python
D1="I like machine learning"
D2="I like deep learning"
D3="I like playing tennins"
D4="I like playing cricket"
D1="I like machine learning"
D2="I like deep learning"
D3="I like playing tennis"
D4="I like playing cricket"
D5="I like data science"
D6="I like artificial intelligence"
D7="I like neural networks"
D8="I like natural language processing"
D9="I like computer vision"
D10="I like reinforcement learning"
D11="I like reading books"
D12="I like watching movies"
D13="I like listening to music"
D14="I like coding in Python"
D15="I like web development"
D16="I like mobile app development"
D17="I like cloud computing"
D18="I like big data analytics"
D19="I like playing football"
D20="I like playing basketball"
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# List of documents (make sure D1â€“D20 are defined as strings)
docs = [D1, D2, D3, D4, D5, D6, D7, D8, D9, D10,
        D11, D12, D13, D14, D15, D16, D17, D18, D19, D20]

# Create CountVectorizer object
cv = CountVectorizer(
    stop_words='english',
    max_df=0.95,
    min_df=1
)

# Generate document-term matrix
dtm = cv.fit_transform(docs)

# Extract vocabulary (feature names)
words = cv.get_feature_names_out()

# Convert sparse matrix to DataFrame
bow_df = pd.DataFrame(dtm.toarray(), columns=words)

# Optional: Add document labels
bow_df.index = [f'Doc{i+1}' for i in range(len(docs))]

print("Bag of Words Representation:")
print(bow_df)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Compute cosine similarity
cos_sim = cosine_similarity(bow_df.values)

# Convert directly to DataFrame using existing index (if already set)
cosine_df = pd.DataFrame(cos_sim,
                         index=bow_df.index,
                         columns=bow_df.index)

print("Cosine Similarity Matrix:")
print(cosine_df.round(3))   # rounded for better readability
```

Jaccard Similarity

```python
from sklearn.metrics import pairwise_distances
import pandas as pd

# Convert BOW to binary (presence = 1, absence = 0)
binary_bow = (bow_df > 0).astype(int)

# Compute Jaccard distance matrix
jaccard_dist = pairwise_distances(binary_bow.values, metric='jaccard')

# Convert distance to similarity
jaccard_sim = 1 - jaccard_dist

# Create DataFrame
doc_labels = [f'Doc{i+1}' for i in range(len(binary_bow))]

df_jaccard = pd.DataFrame(jaccard_sim,
                          index=doc_labels,
                          columns=doc_labels)

print("Jaccard Similarity Matrix:")
print(df_jaccard.round(3))
```

 WordNet-based Similarity

```python
import nltk
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize
import numpy as np
import pandas as pd

try:
    nltk.data.find('corpora/wordnet')
except LookupError:
    nltk.download('wordnet')
try:
    nltk.data.find('tokenizers/punkt_tab')
except LookupError:
    nltk.download('punkt_tab')

def document_wordnet_similarity(doc1, doc2):
    tokens1 = word_tokenize(doc1.lower())
    tokens2 = word_tokenize(doc2.lower())

    synsets1 = [s for token in tokens1 for s in wordnet.synsets(token)]
    synsets2 = [s for token in tokens2 for s in wordnet.synsets(token)]

    if not synsets1 or not synsets2:
        return 0.0

    # Calculate pairwise similarity and take the maximum for each pair
    max_similarities = []
    for s1 in synsets1:
        max_sim_for_s1 = 0.0
        for s2 in synsets2:
            sim = s1.path_similarity(s2)
            if sim is not None and sim > max_sim_for_s1:
                max_sim_for_s1 = sim
        max_similarities.append(max_sim_for_s1)

    # Average the maximum similarities (simple approach)
    if max_similarities:
        return np.mean(max_similarities)
    else:
        return 0.0

# Prepare documents list
documents = [D1, D2, D3, D4, D5, D6, D7, D8, D9, D10, D11, D12, D13, D14, D15, D16, D17, D18, D19, D20]

# Calculate WordNet similarity matrix
num_documents = len(documents)
wordnet_sim_matrix = np.zeros((num_documents, num_documents))

for i in range(num_documents):
    for j in range(num_documents):
        if i == j:
            wordnet_sim_matrix[i, j] = 1.0
        else:
            wordnet_sim_matrix[i, j] = document_wordnet_similarity(documents[i], documents[j])

df_wordnet_sim = pd.DataFrame(wordnet_sim_matrix, index=documents, columns=documents)

print("WordNet Similarity Matrix (Path Similarity):")
print(df_wordnet_sim)
```
