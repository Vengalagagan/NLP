# **Import Libraries**

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

# **Dataset**

```python
df=pd.read_csv('/content/IMDB Dataset.csv', engine='python', on_bad_lines='skip')
df['sentiment'] = df['sentiment'].map({'positive': 1, 'negative': 0})
df.head()
```

# **Feature Extraction (Bag-of-Words)**

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['review'])

y = df['sentiment']
```

# **Data Split**

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

# **Train Naive Bayes Model**

```python
model = MultinomialNB()
model.fit(X_train, y_train)
```

# **Model Testing**

```python
y_pred = model.predict(X_test)

print("Predictions:", y_pred)
print("Actual:", y_test.values)
```

# **Model Evaluation**

```python
print("Accuracy:", accuracy_score(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

# **Model Saving**

```python
import joblib

# Save trained model
joblib.dump(model, 'naive_bayes_model.pkl')

# Save vectorizer also (VERY IMPORTANT)
joblib.dump(vectorizer, 'vectorizer.pkl')
```

# **Load Model**

```python
# Load model
loaded_model = joblib.load('naive_bayes_model.pkl')

# Load vectorizer
loaded_vectorizer = joblib.load('vectorizer.pkl')
```

# **Model Deployment**

```python
text = ["Absolutely loved every second watching"]

# Transform using saved vectorizer
text_vec = loaded_vectorizer.transform(text)

# Predict
prediction = loaded_model.predict(text_vec)

print("Prediction:", prediction[0])
```

# ***Case Study - 2 : News Group Dataset***

# **Import Libraries**

```python
import numpy as np
import pandas as pd

# Sklearn modules
from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

# Evaluation metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

# **Load Dataset**

```python
# Load dataset (subset for binary classification)
categories = ['sci.space', 'rec.sport.hockey']
data = fetch_20newsgroups(subset='all', categories=categories)

texts = data.data
labels = data.target


df_news = pd.DataFrame({'text': texts, 'label': labels})
df_news.head()
```

# **Text Preprocessing**

```python
import re

def preprocess(text):
    text = text.lower()  # lowercase
    text = re.sub(r'[^a-z\s]', '', text)  # remove punctuation
    return text

texts = [preprocess(t) for t in texts]
```

# **Feature Extraction**

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(texts)
```

# **Data Split**

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, labels, test_size=0.2, random_state=42
)
```

# **Model Training**

```python
model = MultinomialNB()
model.fit(X_train, y_train)
```

# **Model Testing**

```python
y_pred = model.predict(X_test)

print("Predictions:", y_pred)
print("Actual:", y_test)
```

# **Model Evaluation**

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

# **Model Saving**

```python
import joblib

# Save trained model
joblib.dump(model, 'naive_bayes_model_news.pkl')

# Save vectorizer also (VERY IMPORTANT)
joblib.dump(vectorizer, 'vectorizer_news.pkl')
```

# **Load Model**

```python
# Load model
loaded_model = joblib.load('naive_bayes_model_news.pkl')

# Load vectorizer
loaded_vectorizer = joblib.load('vectorizer_news.pkl')
```

# **Model Deployment**

```python
test_samples = [
    "The stock market saw a significant rise in tech shares today",
    "Scientists discovered a new species of marine life in the Pacific Ocean",
    "The football match ended in a dramatic penalty shootout",
    "Advancements in artificial intelligence are transforming healthcare",
    "The movie received critical acclaim for its storytelling and visuals"
]

# Transform
test_vec = loaded_vectorizer.transform(test_samples)

# Predict
predictions = loaded_model.predict(test_vec)

# Map labels to category names
for text, pred in zip(test_samples, predictions):
    print("\nText:", text)
    print("Predicted Category:", data.target_names[pred])
```
